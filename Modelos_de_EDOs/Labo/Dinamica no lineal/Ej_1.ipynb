```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tam_paso = [0.01, 0.1, 0.25, 0.5] 
lambdas = [-5, -1, 1, 5]

for l in lambdas:
    print(f'Lambda: {l}')
    
    for h in tam_paso:
        print(f'Tamaño del paso: {h}')
        
        f = lambda t, y: l * y      # EDO    
        t0, y0 = 0, 1               # Condiciones iniciales
        tf = 1
        
        n_pasos  = int((tf-t0) / h)        
        print(f'Número de pasos: {n_pasos}')    
        
        t = np.arange(t0, tf + h, h)
        y = np.zeros(len(t))
        y[0] = y0
        
        for i in range(len(t) - 1):
            y[i+1] = y[i] + h * f(t[i], y[i])
        
        # Gráfico

        plt.figure()
        plt.plot(t, y, 'o-', label='Método de Euler')
        plt.plot(t, np.exp(l * t), label='Solución exacta')
        plt.title(f'λ={l}, h={h}')
        plt.legend()
        plt.grid(True)
        plt.show()
```
