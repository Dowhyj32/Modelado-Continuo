```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import sympy as sp
```

```python
# Ejercicio 8 (Sistema predador-presa)

# x(t): población predadores
# y(y): población presas
# alpha: tasa
# beta: tasa 
# gama:
# delta

X, Y = sp.symbols('X Y')
alpha, beta, gamma, delta = sp.symbols('alpha beta gamma delta')

eq1 = -alpha * X + gamma * X * Y
eq2 = beta * Y - delta * X * Y

equilibrio = sp.solve((eq1, eq2), (X, Y))
print(f'Equilibrio: {equilibrio}')

equilibrio
```

```python
def sistema(t, estado, alpha, beta, gamma, delta):
    x, y = estado 
    dxdt = -alpha * x + gamma * x * y
    dydt = beta * y - delta * x * y

    return [dxdt, dydt]
```

```python
def runge_kutta2(alpha, beta, gamma, delta, x0, y0, t0, tf, h):
    t_valores = [t0]
    predadores_valores = [x0]
    presa_valores = [y0]

    t = t0
    estado = [x0, y0]

    while t < tf:
        # k1 = pendiente al inicio del intervalo
        k1 = sistema(t, estado, alpha, beta, gamma, delta)

        # Estado aproximado en la mitad del paso
        estado_medio = [
            estado[0] + (h / 2) * k1[0],
            estado[1] + (h / 2) * k1[1]
        ]

        # k2 = pendiente en el punto medio
        k2 = sistema(t + h / 2, estado_medio, alpha, beta, gamma, delta)

        # Actualización del estado
        estado = [
            estado[0] + h * k2[0],
            estado[1] + h * k2[1]
        ]

        t = t + h

        t_valores.append(t)
        predadores_valores.append(estado[0])
        presa_valores.append(estado[1])

    return np.array(t_valores), np.array(predadores_valores), np.array(presa_valores)
```

```python
# Todos los parámetros en 0.5

tiempos, predadores, presa = runge_kutta2(0.5, 0.5, 0.5, 0.5, 80, 30, 0, 100, 0.01)

plt.figure(figsize=(15,8))

plt.plot(tiempos, predadores, label='Predadores')
plt.plot(tiempos, presa, label='Presa')
plt.grid(True)
plt.xlabel('t')
plt.ylabel('Población(t)')
plt.title(f'Modelo con α:0.5, β:0.5,  γ:0.5,  δ:0.5')

plt.show()
```

```python
# Parámetros
alpha = 0.5
beta = 0.5
gamma = 0.5
delta = 0.5


# --------------------------------------------------
# Calcular equilibrios con Sympy
# --------------------------------------------------
X, Y = sp.symbols('X Y')
a, b, g, d = sp.symbols('a b g d')

eq1 = -a * X + g * X * Y
eq2 = b * Y - d * X * Y

equilibrios = sp.solve((eq1, eq2), (X, Y))

valores = {a: alpha, b: beta, g: gamma, d: delta}
equilibrios_num = [(float(x.subs(valores)), float(y.subs(valores))) for x, y in equilibrios]

print("Equilibrios:", equilibrios_num)

# Campo vectorial
def f(x, y):
    return -alpha*x + gamma*x*y

def g(x, y):
    return beta*y - delta*x*y

# Grilla del plano fase
x_vals = np.linspace(0, 400, 20)
y_vals = np.linspace(0, 250, 20)
X, Y = np.meshgrid(x_vals, y_vals)

# Componentes del campo
U = f(X, Y)
V = g(X, Y)

norma = np.sqrt(U**2 + V**2)
norma[norma == 0] = 1

U_norm = U / norma
V_norm = V / norma


plt.figure(figsize=(10, 6))
plt.quiver(X, Y, U_norm, V_norm)

# Trayectoria numérica
plt.plot(predadores, presa, color='red', linewidth=2, label='Trayectoria')

# Condición inicial
x0 = 80
y0 = 30
plt.scatter(x0, y0, color='blue', s=80, label='Condición inicial')

# Equilibrios
for i, (x_eq, y_eq) in enumerate(equilibrios_num):
    plt.scatter(x_eq, y_eq, color='red', s=100, label='Equilibrio' if i == 0 else None)


plt.xlabel('Predadores x')
plt.ylabel('Presas y')
plt.title(f'Diagrama de fases - con  α:{alpha},  β:{beta},  γ:{gamma},  δ:{delta}')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Inciso c

# alpha: 0.25
# beta: 1
# gamma = delta = 0.01
# xo=80
# y0=30

# Parámetros

alpha = 0.25
beta = 1
gamma, delta = 0.01, 0.01

# Condiciones iniciales
x0 = 100
y0 = 50
t0 = 0
tf = 20
h = 0.1

# --------------------------------------------------
# Calcular equilibrios con Sympy
# --------------------------------------------------
X, Y = sp.symbols('X Y')
a, b, g, d = sp.symbols('a b g d')

eq1 = -a * X + g * X * Y
eq2 = b * Y - d * X * Y

equilibrios = sp.solve((eq1, eq2), (X, Y))

valores = {a: alpha, b: beta, g: gamma, d: delta}
equilibrios_num = [(float(x.subs(valores)), float(y.subs(valores))) for x, y in equilibrios]

print("Equilibrios:", equilibrios_num)

# Campo vectorial 

def f(X,Y):
    return -alpha * X + gamma * X * Y

def g(X,Y):
    return beta * Y - delta * X * Y

x_vals = np.linspace(0, 800, 25)
y_vals = np.linspace(0, 600, 25)
X, Y = np.meshgrid(x_vals, y_vals) 

U = f(X, Y)
V = g(X, Y)

norma = np.sqrt(U**2 + V**2)
norma[norma==0] = 1

U_norm = U / norma
V_norm = V / norma

# Desarrollo sistema
tiempos, predadores, presa = runge_kutta2(0.25, 1, 0.01, 0.01, x0, y0, t0, tf, h)

# Evolución variables
plt.figure(figsize=(15,8))
plt.scatter(tiempos, predadores, label='Predadores')
plt.scatter(tiempos, presa, label='Presa')
plt.legend()
plt.grid(True)
plt.xlabel('t')
plt.ylabel('Población(t)')
plt.title(f'Modelo con α:0.25, β:1,  γ:0.01,  δ:0.01. Inicialmente: depredadores={x0}; presa={y0}')
plt.show()



# Diagrama de fases
plt.figure(figsize=(10, 6))
plt.quiver(X, Y, U_norm, V_norm)

condiciones_iniciales = [[100,50],[50,100],[200,50],[50,200],[500,300]]

for estado in condiciones_iniciales:
    tiempos, predadores, presa = runge_kutta2(0.25, 1, 0.01, 0.01, estado[0], estado[1], t0, tf, h)
    plt.plot(predadores, presa)
    plt.plot(estado[0], estado[1], 'ro', markersize=6, label=f'x0: {estado[0]}  y0: {estado[1]}', color='y')
    
plt.plot(0, 0, 'ro', markersize=6, label='Equilibrio', color='red')
plt.plot(100, 25, 'ro', markersize=6, color='red')
plt.title(f'Diagrama de fases (depredadores vs presa) en un período de {tf}')
plt.legend()
plt.show()
```
