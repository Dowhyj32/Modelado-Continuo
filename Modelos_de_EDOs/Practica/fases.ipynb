```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def sistema(estado):
    x, y = estado
    dXdt = x * (1-x)
    dYdt = y-2
    return np.array([dXdt, dYdt])

def euler(x0, y0, t0, tf, h):
    num_pasos = int((tf-t0)/h)
    t = np.linspace(t0, tf, num_pasos+1)
    
    X = np.zeros(num_pasos+1)
    X[0] = x0
    Y = np.zeros(num_pasos+1)
    Y[0] = y0 
    
    for i in range(num_pasos):
        estado = np.array([X[i], Y[i]])
        pendiente = sistema(estado)
        
        X[i+1] = X[i] + h * pendiente[0]
        Y[i+1] = Y[i] + h * pendiente[1]
        
    return t, X, Y
```

```python
# Condiciones iniciales
condiciones_iniciales = [
    (0.2, 2.1),
    (-0.5, 1),
    (-0.5, 3),
    (0.2, 1),
    (0.2, 3),
    (0.8, 1),
    (0.8, 3),
    (1.5, 1),
    (1.5, 3),
    (2.5, 1),
    (2.5, 3)
]

# Campo vectorial
x_vals = np.linspace(-1, 3, 25)
y_vals = np.linspace(0, 4, 25)
X_grid, Y_grid = np.meshgrid(x_vals, y_vals)

U = X_grid * (1 - X_grid)
V = Y_grid - 2

norma = np.sqrt(U**2 + V**2)
norma[norma == 0] = 1

U_norm = U / norma
V_norm = V / norma

# Gráfico
plt.figure(figsize=(10, 7))
plt.quiver(X_grid, Y_grid, U_norm, V_norm, alpha=0.6)

# Equilibrios
plt.scatter(0, 2, color='red', s=80, label='(0,2)')
plt.scatter(1, 2, color='blue', s=80, label='(1,2)')

# Trayectorias
for x0, y0 in condiciones_iniciales:
    t, x, y = euler(x0, y0, 0, 30, 0.01)
    plt.plot(x, y, linewidth=2)

plt.axvline(0, linestyle='--', alpha=0.5)
plt.axvline(1, linestyle='--', alpha=0.5)
plt.axhline(2, linestyle='--', alpha=0.5)

plt.xlim(-1, 3)
plt.ylim(0, 4)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo vectorial y trayectorias')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
